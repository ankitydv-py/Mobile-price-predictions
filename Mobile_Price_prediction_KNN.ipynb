# KNN Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
training_dataset = pd.read_csv("mobile_train.csv")
```

```python
training_dataset.head()
```

```python
training_dataset.info()
```

```python
#check for the null values in the dataset
training_dataset.isnull().values.any()
```

```python
#set the price_range as the target value

target_value = training_dataset.price_range
```

```python
training_dataset.describe()
```

```python
#drop the attributes bluetooth, mobile weight, talktime, wifi and prince range(target) 

training_dataset_x = training_dataset.drop(['price_range','blue','mobile_wt','talk_time','wifi'],axis = 'columns')
```

```python
training_dataset_x.info()
```

```python
#split dataset into train and testing datasets
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(training_dataset_x, target_value, test_size=0.33, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=10)
knn_model.fit(X_train,Y_train)
```

```python
knn_model.score(X_test,Y_test)
```

```python
#now prediction of the test dataset
```

```python
testing_dataset = pd.read_csv("mobile_test.csv")
```

```python
testing_dataset.info()
```

```python
#drop the attributes id,bluetooth, mobile weight, talktime, wifi
```

```python
testing_dataset_x = testing_dataset.drop(['id','blue','mobile_wt','talk_time','wifi'],axis = 'columns')
```

```python
knn_model.predict(testing_dataset_x)
```
