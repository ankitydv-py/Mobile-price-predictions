```python
train_dataset_x = dataset.drop(['price_range','blue','mobile_wt','talk_time','wifi'],axis = 'columns')
```

```python
#split dataset into train and testing datasets
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train_dataset_x, target, test_size=0.33, random_state=101)
```

```python
#model training
```

```python
from sklearn.linear_model import LinearRegression
pred_model = LinearRegression()
```

```python
pred_model.fit(X_train,Y_train)
```

```python
pred_model.score(X_test,Y_test)
```
